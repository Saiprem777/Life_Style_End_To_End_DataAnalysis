```python
df.drop(['BMI_calc','cal_from_macros','pct_carbs',
 'protein_per_kg',
 'pct_HRR',
 'pct_maxHR',
 'cal_balance',
 'lean_mass_kg',
 'expected_burn',
 'Burns Calories (per 30 min)_bc','sugar_g',
 'sodium_mg',
 'cholesterol_mg',
 'serving_size_g', 'prep_time_min',
 'cook_time_min',
 'rating','Benefit','Daily meals frequency',
 'Physical exercise','meal_name',
 'Burns Calories (per 30 min)'],axis = 1 , inplace = True)
```

```python
df.columns.tolist()
```

```python
df.rename(columns={'Fat_Percentage':'Body_Fat_Percentage',
                   'Carbs':'Carbs_Intake',
                   'Proteins' : 'Protein_Intake',
                   'Fats': 'Facts_Intake',
                   'Calories' : 'Calories_Intake',
                   'meal_type': 'Meal_type',
                   'diet_type' : 'Diet_type',
                   'cooking_method' : 'Cooking_method',
                   'Name of Exercise':'Name_of_Exercise',
                   'Target Muscle Group':'Target_Muscle_Group',
                   'Equipment Needed':'Equipment_Needed',
                   'Difficulty Level': 'Difficulty_Level',
                    'Body Part': 'Body_Part',
                    'Type of Muscle' :'Type_of_Muscle'},inplace= True)
```

```python
df.columns.tolist()
```

```python
df.rename(columns={'Weight (kg)':'Weight_Kg',
                  'Height (m)':'Height_m',
                   'Session_Duration (hours)':'Session_Duration_hours',
                   'Water_Intake (liters)':'Water_Intake_liters',
                   'Sleep_Duration (hours)':'Sleep_Duration_hours',
 'Workout_Frequency (days/week)':'Workout_Frequency_Days_Per_Week'},inplace= True)
```

```python
df['Age'] = df['Age'].astype(int)

df['Session_Duration_hours'] = (df['Session_Duration_hours'] * 60).astype(int)

df['Workout_Frequency_Days_Per_Week'] = df['Workout_Frequency_Days_Per_Week'].round(0).astype(int)


df['Experience_Level_Int'] = df['Experience_Level'].round(0).astype(int)
df['Experience_Level'] = df['Experience_Level_Int'].map({
    1: 'Beginner',
    2: 'Intermediate',
    3: 'Advanced'
})

# Optional: drop the temporary column
df.drop(columns=['Experience_Level_Int'], inplace=True)


df['Sets'] = df['Sets'].round(0).astype(int)
df['Reps'] = df['Reps'].round(0).astype(int)


print(df[['Age', 'Session_Duration_hours', 'Workout_Frequency_Days_Per_Week', 'Experience_Level', 'Sets', 'Reps']].head())
```
